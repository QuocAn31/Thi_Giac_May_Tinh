```python
# Import lỚP iMAGE
from PIL import Image
from imgtools import *
from IPython.display import display # to
```

```python
# Đường dẫn ảnh
dir = 'C:/Picture'
image_path = dir + '/1.jpg'
print(image_path)
```

```python
img = load_image(image_path)
display(img)
```

```python
# Định vị khu vực cắt 
region = (100, 75, 175, 160)
# Cắt vùng ảnh
cropped_image = img.crop(region)
display(cropped_image)
```

```python
# Dán
# Xác định vị trí cần dán
paste_position = (150,150)
img.paste(cropped_image, paste_position)
display(img)
```
